### 예시 돌려보기

```python
import base64
import vertexai
from vertexai.generative_models import GenerativeModel, Part, Image, FinishReason
import vertexai.preview.generative_models as generative_models
from IPython.display import Video
```

```python
def generate(video, text):
  vertexai.init(project="test-gemini-421322", location="us-central1")
  model = GenerativeModel("gemini-1.0-pro-vision-001")
  responses = model.generate_content(
      [video, text],
      generation_config=generation_config,
      safety_settings=safety_settings,
      stream=True,
  )

  for response in responses:
    print(response.text, end="")

video1 = Part.from_data(
    mime_type="video/mp4",
    data=base64.b64decode(""""""))

generation_config = {
    "max_output_tokens": 2048,
    "temperature": 0.4,
    "top_p": 1,
    "top_k": 32,
}

safety_settings = {
    generative_models.HarmCategory.HARM_CATEGORY_HATE_SPEECH: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_HARASSMENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
}
```

```python
generate(video=video1, text="이 영상 설명해줘")
```

## Local File 요청

```python
Video("3327172-hd_1920_1080_24fps.mp4", width=640)
```

1. 파일을 binary 파일로 열어서 `f.read()`로 읽기
2. Video encode
3. `from_data()`에는 문자열 형식으로 넣어야 하기 때문에 utf-8로 decode

```python
with open("3327172-hd_1920_1080_24fps.mp4", "rb") as f:
    bin = f.read()
```

```python
video_encoded = base64.b64encode(bin)
```

```python
video2 = Part.from_data(data=video_encoded.decode('utf-8'), mime_type="video/mp4")
```

```python
generate(video2, "이 영상을 설명해줘")
```
